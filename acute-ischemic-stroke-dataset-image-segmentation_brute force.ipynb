```python
import os
import tarfile

def extract_tar_gz(file_path, extract_to=None):
    """
    Mengekstrak file .tar.gz

    Parameters:
    - file_path: path ke file .tar.gz
    - extract_to: folder tujuan ekstraksi (default: folder yang sama dengan file)
    """
    if not os.path.isfile(file_path):
        raise FileNotFoundError(f"File tidak ditemukan: {file_path}")
    
    if extract_to is None:
        extract_to = os.path.dirname(file_path)

    os.makedirs(extract_to, exist_ok=True)

    try:
        with tarfile.open(file_path, "r:gz") as tar:
            tar.extractall(path=extract_to)
        print(f"Berhasil ekstrak: {file_path} → {extract_to}")
    except tarfile.TarError as e:
        print(f"Gagal mengekstrak {file_path}: {e}")
```

```python
target_dir = "/kaggle/working/Acute-ischemic-stroke-dataset-image-segmentation/"
for file_, id_drive in dicom_drive_dict.items():
    extract_tar_gz(file_, target_dir)
```

```python
import os

all_patient_dicom_data = []
for file_, id_drive in dicom_drive_dict.items():
    complete_path = os.path.join(target_dir, file_.replace(".tar.gz",""))
    all_patient_id = [os.path.join(complete_path, i) for i in os.listdir(complete_path)]
    all_patient_dicom_data.extend(all_patient_id)
```

```python
import csv
import subprocess
from pathlib import Path

import numpy as np

# Resume / skip mechanism:
# - Jika kamu upload CSV sebelumnya (mis. `lesion_patients.csv`), semua patient_id yang sudah ada
#   di CSV itu akan di-skip.
resume_csv_path = "lesion_patients.csv"  # ganti kalau nama file berbeda

# Optional uploader (aktifkan jika environment mendukung, mis. Google Colab)
# try:
#     from google.colab import files
#     uploaded = files.upload()  # pilih CSV resume
#     if uploaded:
#         resume_csv_path = next(iter(uploaded.keys()))
# except Exception:
#     pass

processed_ids = set()
prior_rows = []
if resume_csv_path and Path(resume_csv_path).exists():
    with open(resume_csv_path, "r", encoding="utf-8") as f:
        r = csv.DictReader(f)
        for row in r:
            pid = (row.get("patient_id") or "").strip()
            if pid:
                processed_ids.add(pid)
                prior_rows.append(row)
    print(f"Resume aktif: skip {len(processed_ids)} patient_id dari {resume_csv_path}.")
else:
    print("Resume non-aktif (CSV tidak ditemukan).")

lesion_hits = []
failed = []

outputs_root = Path("outputs_dicom")
outputs_root.mkdir(parents=True, exist_ok=True)

for data in all_patient_dicom_data:
    patient_id = str(data).split("/")[-1]
    if patient_id in processed_ids:
        print(f"{patient_id}: SKIP")
        continue

    dicom_ct_path = f"{data}/CT"
    out_dir = outputs_root / f"{patient_id}_CT"

    cmd = [
        "python",
        "infer_dicom_unet.py",
        "--dicom",
        dicom_ct_path,
        "--ckpt",
        "best_unet.pt",
        "--out-dir",
        str(out_dir),
    ]

    try:
        subprocess.run(cmd, check=True, capture_output=True, text=True)

        mask_path = out_dir / "mask_pred.npy"
        if not mask_path.exists():
            failed.append({"patient_id": patient_id, "dicom_ct_path": dicom_ct_path, "reason": "mask_pred.npy not found"})
            continue

        m = np.load(mask_path)
        lesion_voxels = int(m.sum())

        if lesion_voxels > 0:
            lesion_hits.append(
                {
                    "patient_id": patient_id,
                    "dicom_ct_path": dicom_ct_path,
                    "out_dir": str(out_dir),
                    "lesion_voxels": lesion_voxels,
                }
            )

        print(f"{patient_id}: lesion_voxels={lesion_voxels}")

    except subprocess.CalledProcessError as e:
        failed.append({"patient_id": patient_id, "dicom_ct_path": dicom_ct_path, "reason": (e.stderr or "CalledProcessError")})
        print(f"{patient_id}: FAILED")

# Gabungkan prior rows + hits baru, dan simpan kembali
merged = []
for row in prior_rows:
    pid = (row.get("patient_id") or "").strip()
    if not pid:
        continue
    merged.append(
        {
            "patient_id": pid,
            "dicom_ct_path": row.get("dicom_ct_path", ""),
            "out_dir": row.get("out_dir", ""),
            "lesion_voxels": int(float(row.get("lesion_voxels") or 0)),
        }
    )
merged.extend(lesion_hits)

csv_path = Path("lesion_patients.csv")
with csv_path.open("w", newline="", encoding="utf-8") as f:
    w = csv.DictWriter(f, fieldnames=["patient_id", "dicom_ct_path", "out_dir", "lesion_voxels"])
    w.writeheader()
    w.writerows(merged)

failed_csv_path = Path("lesion_patients_failed.csv")
with failed_csv_path.open("w", newline="", encoding="utf-8") as f:
    w = csv.DictWriter(f, fieldnames=["patient_id", "dicom_ct_path", "reason"])
    w.writeheader()
    w.writerows(failed)

print(f"Selesai. Hits baru: {len(lesion_hits)}. Total hits: {len(merged)}. CSV: {csv_path}")
print(f"Gagal diproses run ini: {len(failed)} pasien. CSV: {failed_csv_path}")
```
